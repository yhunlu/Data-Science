## Manipulating Data 2

```python
# column from series
seats_column = pd.Series([5, 5, 5, 5, 5])

# new colum called seats
car_sales["Seats"] = seats_column
car_sales
```

```python
# fill 5 for Nan
car_sales["Seats"].fillna(5, inplace=True)
car_sales
```

```python
# column from python list
# list lenght must match with table's row lenght
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5, 5]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
# add new columns as calculation result.
car_sales["Total fuel used"] = car_sales["Odometer (KM)"] /100 * car_sales["Fuel per 100KM"]
```

```python
car_sales
```

```python
# Create a column from as single value
car_sales
```

```python
car_sales["Passed road saftey"] = True
```

```python
car_sales.dtypes
```

```python
car_sales
```

```python
# remove column.
car_sales.drop("Total fuel used", axis=1)
```
