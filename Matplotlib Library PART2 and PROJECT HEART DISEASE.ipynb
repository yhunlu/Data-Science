### Making figures with numPy array

list of baseplots:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
# create sample data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**3)
```

```python
# scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# bar plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title='Nut Butter Store',
      ylabel='Price ($)');
```

```python
# horizontal bar plot
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# histogram plot
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

#### pyplot vs matplotlib

* When plotting something quickly, use pyplot "pandas"
* When plotting something more advanced, use OO method "matplotlib"

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
# pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# matplotlib OO method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
# ax.set_xlim([45, 100]);
```

```python
over_30 = heart_disease[heart_disease['age'] > 30]
```

```python
fix, ax = plt.subplots(figsize=(10, 6))
over_30.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

### OO method with customize

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Plot data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target']);
# Customize plot
ax.set(title='Heart Disease and Cholesterol Levels',
      xlabel='Age',
      ylabel='Cholesterol');
# Add legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# Subplots of chol, age, thalach
fig, ax = plt.subplots(nrows=2,
                      ncols=1,
                      figsize=(10, 10),
                      sharex=True)
# add data to axis
ax[0].scatter(x=over_50['age'],
                y=over_50['chol'],
                c=over_50['target']);

ax[1].scatter(x=over_50['age'],
                y=over_50['thalach'],
                c=over_50['target']);

# Customize axis
ax[0].set(title='Heart Disease and Cholesterol Levels',
         ylabel='Cholesterol');

ax[1].set(title='Heart Disease and Heart-Rate Levels',
         xlabel='Age',
         ylabel='Heart-Rate');

# add legend to axis
ax[0].legend(*scatter.legend_elements(), title='Target');

ax[1].legend(*scatter.legend_elements(), title='Target');

# add meanline
ax[0].axhline(y=over_50['chol'].mean(),
             linestyle='--');

ax[1].axhline(y=over_50['thalach'].mean(),
             linestyle='--');

# Add title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

#### HEART DISEASE GRAPHICAL DATA

```python
# assign style
plt.style.use('seaborn')

# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                      ncols=1,
                      figsize=(10, 10),
                      sharex=True)
# Plot data
scatter = ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter');

scatter = ax1.scatter(x=over_50['age'],
                    y=over_50['thalach'],
                    c=over_50['target'],
                    cmap='winter');

# Customize plot
ax0.set(title='Heart Disease and Cholesterol Levels',
      ylabel='Cholesterol');

ax1.set(title='Heart Disease and Max Heart Rate',
      xlabel='Age',
      ylabel='Max Heart Rate');

# x axis limits
ax0.set_xlim([50, 80])

# y axis limits
ax1.set_ylim([60, 200])

# Add legend
ax0.legend(*scatter.legend_elements(), title='Target');
ax1.legend(*scatter.legend_elements(), title='Target');

# Add horizontal line
ax0.axhline(y=over_50['chol'].mean(),
          linestyle='--',
          c='darkorange');

ax1.axhline(y=over_50['thalach'].mean(),
          linestyle='--',
          c='darkorange');

# Add title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
